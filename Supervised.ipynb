```python
import pandas as pd
```

```python

heart_disease=pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
```

```python
heart_disease
```

```python
heart_disease.head()
```

```python
heart_disease.info
```

```python
X= heart_disease.drop("target",axis=1)
X
```

```python
Y=heart_disease.target
Y
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.25)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
```

```python
clf.get_params()
```

```python
clf.fit(X=X_train,y=y_train)
```

```python
clf.score(X,Y)
```

```python
X_test.head()
```

```python
X_test.tail()
```

```python
#Doing the prediction
y_preds=clf.predict(X_test)
```

```python
#Evaluating the model
clf.score(X_train,y_train)
train=clf.score(X_train,y_train)
print(f"The accuracy of the training is:  {train*100:.2f}%")
```

```python
test=clf.score(X_test,y_test)
print(f"The accuracy of the test is: {test*100:.2f}%")
```

```python
test
```

```python

predo=clf.score(X_test,y_preds)
print(f"The accuracy of the prediction is {predo*100:.2f}%")
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#Creating a classification report
print(classification_report(y_test,y_preds))
```

```python
#Creating the confusion matrix
confusion_matrix(y_test,y_preds)
```

```python
#accuracy score
accuracy_score(y_test,y_preds)
```

```python
#Experimentation
```

```python
import numpy as np
```

```python
np.random.seed(42)
for i in range(100,200,10):
    print(f"Trying with {i} estimators...")
    model=RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(f"Model accuracy is: {model.score(X_test,y_test)*100:.2f}%")
    print("")
```

```python
#cross validation score
```

```python
from sklearn.model_selection import cross_val_score
```

```python
np.random.seed(42)
for i in range(100,200,10):
    print(f"Trying with {i} estimators...")
    model=RandomForestClassifier(n_estimators=i).fit(X_train,y_train)

    model_score=model.score(X_test,y_test)
    print(f"Model accuracy on test set is: {model_score*100:.2f}%")

    cv_score_mean=np.mean(cross_val_score(model,X,Y,cv=5))
    print(f"5-folds Cross Validation score:{cv_score_mean*100:.2f}%")
    print("")
    
```

```python
np.random.seed(42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={"n_estimators":[ i for i in range(100,200,10)]}
grid=GridSearchCV(estimator=RandomForestClassifier(),
                  param_grid=param_grid,
                  cv=5,
                  verbose=1)
grid.fit(X,Y)
print(f"The best parameters are: {grid.best_params_}")
print(f"with a score of: {grid.best_score_*100:.2f}%")
```

```python
clf=grid.best_estimator_
clf
```

```python
clf.fit(X_train,y_train)
print(f"The best on a single split is {clf.score(X_test,y_test)*100:.2f}%")
```
