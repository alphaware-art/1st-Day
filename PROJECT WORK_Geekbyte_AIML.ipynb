1.PROBLEM DEFINITION:

- Predict the auction sale price of heavy machinery.

- The target variable in this dataset is "saleprice" and it is a regression type machine learning task because I am going to predict continuous numerical values.

2.DATA

- The data

EVALUATION
- The evaluation metric is the Root Mean Squared Log Error(RSMLE)
- The Root Mean Squared Log Error metric is to be used because it measures the ratio between the actual and predicted values. The RSMLE is required in this dataset because if I prredict a price too low for the auction price,the machinery may be sold for a price which is considered too low and the company going to use this model will end up making losses.If I were to predict the saleprice of a model to be be too low, it would significantly reduce the accuracy of the model. This will help because as said earlier, the prices have to be a bit highly estimated in order for the comapny in possession of this model to not make losses. However, care will be taken to ensure that the prices are not too high to scare away customers.

FEATURES
- Year Made
- Product Size
- SaleDate
These features were selected because they represent the physical value and the market value of the machinery

```python
import pandas as pd
```

```python
import numpy as np
```

```python
Dataset = pd.read_csv("TrainandValid.csv",low_memory=False,parse_dates=["saledate"])
```

```python
Dataset
```

```python
Dataset
```

```python
Dataset.sort_values(by = ["saledate"], inplace = True,ascending = True)
```

```python
Dataset
```

```python

Dataset["Year"] = Dataset['saledate'].dt.year
Dataset["Month"] = Dataset['saledate'].dt.month
Dataset["Day"] = Dataset['saledate'].dt.day
```

```python
Dataset
```

```python
Dataset.info()
```

```python
Dataset.dtypes
```

```python
#Converting the string datatypes to categorical datatypes
for col in Dataset.select_dtypes(include=['object']).columns:
        Dataset[col] = Dataset[col].astype('category')
```

```python
Dataset.info()
```

```python
#Identifying the missing values
Dataset.isnull()
```

```python
Dataset.isnull().sum()
```

```python
#Handling Numerical Missing Values
median_A = Dataset["MachineHoursCurrentMeter"].median()
Dataset["MachineHoursCurrentMeter"] = Dataset["MachineHoursCurrentMeter"].fillna(median_A)
median_B = Dataset["auctioneerID"].median()
Dataset["auctioneerID"] = Dataset["auctioneerID"].fillna(median_B)
```

```python
Dataset.info()
```

```python
#Filling the missing categorical values
from pandas.api.types import CategoricalDtype
for label,content in Dataset.items():
 if isinstance(content.dtype,CategoricalDtype):
     Dataset[label] = content.astype(str)
     Dataset[label] = Dataset[label].replace('nan','Missing')
     Dataset[label] =Dataset[label].astype('category')
```

```python
Dataset
```

```python
#Encoding the categorical values
for label,content in Dataset.items():
 if isinstance(content.dtype,CategoricalDtype):
     Dataset[label] = pd.Categorical(content).codes +1
```

```python
Dataset.info()
```

```python
#Training and Validation Split
Dataset_valid = Dataset[Dataset.Year== 2012]
Dataset_train = Dataset[Dataset.Year != 2012]
```

```python
Dataset_valid
```

```python
#Doing the split
X_train = Dataset_train.drop(["SalePrice","saledate"],axis = 1)
y_train = Dataset_train["SalePrice"]
```

```python
X_valid= Dataset_valid.drop(["SalePrice","saledate"],axis = 1)
y_valid = Dataset_valid["SalePrice"]
```

```python
#Fitting the model
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
#Baseline model has been created
```

```python
#Building an Evaluation Function
```

```python
preds = model.predict(X_valid)
```

```python
from sklearn.metrics import root_mean_squared_log_error
rmsle_score = root_mean_squared_log_error(y_valid,preds)
print(f"Final RSMLE: {rmsle_score:.4f}")
```

```python
Dataset_valid.info()
```

OPTIMIZATION STRATEGY.
 - Max_samples can be set to a definite value such as 1000,2000,3000 or any other value. This actually tells scikit learn to go through
   the set number of samples without going through the entire dataset which could negatively affect things when you do not have a large
   memory. In this case,the training will only be done on a subset of the entire dataset.
  


```python
from sklearn.model_selection import RandomizedSearchCV
param_distributions = {
    'n_estimators' :[20,40],
    'max_depth' :[None,10],
    'min_samples_split' : [2,5,10],
    'min_samples_leaf' : [3,5],
    'bootstrap' :[True]
}
    
    
```

```python
rs_model = RandomizedSearchCV(RandomForestRegressor(max_samples = 10000),
                              param_distributions = param_distributions,
                              n_iter = 2,
                              cv = 2,
                              verbose = True,
                             )
rs_model.fit(X_train,y_train)
```

```python
rs_model.best_params_
```

```python
model = RandomForestRegressor(n_estimators = 100,
                                  min_samples_leaf = rs_model.best_params_['min_samples_leaf'],
                                  min_samples_split = rs_model.best_params_['min_samples_split'],
                                  max_samples = 10000)
model.fit(X_train,y_train)
```

```python
print(f"the score of your new model is:{model.score(X_valid,y_valid)*100:.2f}%")
```

```python
new_preds = model.predict(X_valid)
print(f"Final RMSLE: {np.sqrt(root_mean_squared_log_error(y_valid,new_preds))}")
```
